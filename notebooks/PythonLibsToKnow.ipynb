# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(1,10,15)
a
```

```python
b = a ** 2
b
```

```python
plt.plot(a,b)
```

```python
plt.subplot(1,2,1)
plt.plot(a,b)
plt.title("dito training 1")
plt.subplot(1,2,2)
plt.plot(b,a,"r")
plt.title("Dito training 2")
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
a = np.linspace(1,10,15)
b = a ** 2
fig = plt.figure()
axes = fig.add_axes([.1,.1,.8,.8])
axes.plot(a,b)
```

```python
a = np.linspace(1,10,15)
b = a ** 2
fig = plt.figure()
axes1 = fig.add_axes([.1,.1,.8,.8])
axes2 = fig.add_axes([.2,.5,.4,.3])
axes1.plot(a,b)
axes2.plot(b,a)
axes1.set_xlabel("x1 Label")
axes1.set_ylabel("Y1 label")
axes1.set_title("Title 1")
axes2.set_xlabel("X2 Label")
axes2.set_ylabel("y2 label")
axes2.set_title("Title 2")
```

```python
a = np.linspace(1,10,15)
b = a ** 2
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(a, a**2, label = 'a squared')
ax.plot(a, a**3, label = 'a cubed')
ax.legend()
```

# Seaborn

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'], kde=True, bins=40)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip', data=tips, kind='hex')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips,
               hue='sex', split=True)
```
